```python
import ast
def genre_list(list_dict):
    lis=[]
    for dic in ast.literal_eval(list_dict):
        lis.append(dic["name"])
    return lis
        
        
```

```python
import ast
def convert2(list_dict):
    cnt=0
    lis=[]
    for dic in ast.literal_eval(list_dict):
        if cnt!=5:
            lis.append(dic["name"])
        else:
            break
        cnt+=1
    return lis
        
        
```

```python
movies['cast']=movies['cast'].apply(convert2)
movies['cast']
```

```python
def convert3(list_dict):
    import ast
    direct=[]
    for dic in ast.literal_eval(list_dict):
        if dic['job']=="Director":
            direct.append(dic["name"])
            break
    return direct
        
        
```

```python
convert3(movies['crew'][0])
```

```python
movies["crew"]=movies["crew"].apply(convert3)
movies["crew"]
```

```python
def convert_overview(overview):
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize
    stop_words = set(stopwords.words('english'))
    lis=[",","."]
    stop_words.update(lis)
    word_tokens = word_tokenize(overview)
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    return filtered_sentence
```

```python
movies["overview"]=movies['overview'].apply(convert_overview)
movies["overview"]
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    lis=[]
    for i in text.split():
        lis.append(ps.stem(i))
    return " ".join(lis)
```

```python
new_df["tags"].apply(stem)
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000,stop_words='english')
vectors=cv.fit_transform(new_df["tags"]).toarray()
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
    movie_index=new_df[new_df["title"]==movie].index[0]
    distances=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])
    for movie in distances[1:6]:
        print(new_df.iloc[movie[0]].title)
```

```python
recommend('Batman Begins')
```
